```python
clean_df = clean_df.rename(columns={
    "ticket_no":"ticket_number",
    "between":"between_street_1",
    "and": "between_street_2",
    "from_date_time": "ticket_start_time",
    "to_date_time": "ticket_end_time",
    "amount": "ticket_amount"
})
```

```python
clean_df
```

```python
clean_df["ticket_start_time"] = pd.to_datetime(
    clean_df["ticket_start_time"], errors ="coerce"
)

clean_df["ticket_end_time"] = pd.to_datetime(
    clean_df["ticket_end_time"], errors ="coerce"
)
```

```python
clean_df
```

```python
clean_df["ticket_year"] = clean_df["ticket_start_time"].dt.year
clean_df["ticket_month"] = clean_df["ticket_start_time"].dt.month
clean_df["ticket_day"] = clean_df["ticket_start_time"].dt.day
clean_df["ticket_day_name"]= clean_df["ticket_start_time"].dt.day_name()
clean_df["ticket_hour"]= clean_df["ticket_start_time"].dt.hour
clean_df["is_weekend"]= clean_df["ticket_start_time"].dt.dayofweek >=5
```

```python
clean_df
```

```python
# clean_df[["latitude","longitude"]]= clean_df["gps"].str.split(";", expand= True)
# clean_df["latitude"] = pd.to_numeric(clean_df["latitude"], errors ="coearce")
# clean_df["longitude"]= pd.to_numeric(clean_df["longitude"], errors = "coearce")

clean_df[["latitude", "longitude"]] = clean_df["gps"].str.split(";", expand=True)

clean_df["latitude"] = pd.to_numeric(clean_df["latitude"], errors="coerce")
clean_df["longitude"] = pd.to_numeric(clean_df["longitude"], errors="coerce")
```
